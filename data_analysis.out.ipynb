# PLOT 4 Kmeans clustering for NAICS2_NAME skills (for Analyst position)

```python
from collections import Counter
import ast
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Filter for Analyst positions
df_analyst = df_selected[df_selected['TITLE_CLEAN'].str.contains('analyst', case=False, na=False)].copy()

# Merge skill slots and produce skills_text
def merge_skills_columns_pandas(df):
    def parse_skills(row):
        all_skills = []
        for col in ['COMMON_SKILLS_NAME', 'SPECIALIZED_SKILLS_NAME', 'SOFTWARE_SKILLS_NAME']:
            val = row[col]
            try:
                items = ast.literal_eval(val) if isinstance(val, str) and val.startswith("[") else val.split(';')
                all_skills.extend([skill.strip() for skill in items if skill.strip()])
            except:
                continue
        return ' '.join(all_skills)

    # Filter out information with missing skills
    df_filtered = df.dropna(subset=["COMMON_SKILLS_NAME", "SPECIALIZED_SKILLS_NAME", "SOFTWARE_SKILLS_NAME"]).copy()
    df_filtered['skills_text'] = df_filtered.apply(parse_skills, axis=1)
    return df_filtered

df_skills = merge_skills_columns_pandas(df_analyst)

```

```python
print(df_skills['TITLE_NAME'].unique()) 
```

```python
vectorizer = CountVectorizer(max_features=100)
df_sample = df_skills.sample(n=1000, random_state=42) 
X_sample = vectorizer.fit_transform(df_sample['skills_text']) 
print(vectorizer.get_feature_names_out()[:10])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import MiniBatchKMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

inertias = []
sil_scores = []
K_range = range(2, 6)

for k in K_range:
    mbk = MiniBatchKMeans(n_clusters=k, random_state=42, batch_size=256, n_init=10)
    labels = mbk.fit_predict(X_sample)
    inertias.append(mbk.inertia_)
    sil_scores.append(silhouette_score(X_sample, labels))

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].plot(K_range, inertias, marker='o')
ax[0].set_title("Elbow Method (Inertia)")
ax[0].set_xlabel("Number of Clusters (k)")
ax[0].set_ylabel("Inertia")

ax[1].plot(K_range, sil_scores, marker='o', color='orange')
ax[1].set_title("Silhouette Score")
ax[1].set_xlabel("Number of Clusters (k)")
ax[1].set_ylabel("Score")

plt.tight_layout()
plt.show()

```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(max_features=100)
X = vectorizer.fit_transform(df_skills['skills_text'])
```

```python
from sklearn.cluster import KMeans

best_k = 4
kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)
labels = kmeans.fit_predict(X)  

df_skills['cluster'] = labels  
```

```python
from collections import Counter
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X.toarray())  
df_skills['PCA1'] = X_pca[:, 0]
df_skills['PCA2'] = X_pca[:, 1]


cluster_centers = df_skills.groupby('cluster')[['PCA1', 'PCA2']].mean()
cluster_labels = {}

for c in range(best_k):
    cluster_data = df_skills[df_skills['cluster'] == c]
    skill_words = ' '.join(cluster_data['skills_text']).split()
    top_skills = [w for w, _ in Counter(skill_words).most_common(5)]
    cluster_labels[c] = ', '.join(top_skills)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 8))
sns.scatterplot(data=df_skills, x='PCA1', y='PCA2', hue='cluster', palette='tab10', s=50, alpha=0.6)


for i, row in cluster_centers.iterrows():
    plt.text(row['PCA1'], row['PCA2'], cluster_labels[i], fontsize=10, weight='bold',
             bbox=dict(facecolor='white', alpha=0.8, boxstyle='round,pad=0.3'))

plt.title('Skill-Themed Clusters via PCA', fontsize=16)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.tight_layout()
plt.show()
```

# PLOT 5 Random Forest Classification

```python
import pandas as pd
import numpy as np
import ast
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.metrics import mean_squared_error, r2_score
from scipy.stats import randint

# STEP 1: Drop missing values
df_rf = df_da[[
    "SALARY", "STATE", 
    "COMMON_SKILLS_NAME", "SPECIALIZED_SKILLS_NAME", "SOFTWARE_SKILLS_NAME"
]]

df_rf = df_rf.dropna()

# STEP 2: Combine all skill columns into one clean text field
def combine_skills(row):
    skills = []
    for col in ["COMMON_SKILLS_NAME", "SPECIALIZED_SKILLS_NAME", "SOFTWARE_SKILLS_NAME"]:
        val = row.get(col, "")
        try:
            # Convert stringified list to actual list
            items = ast.literal_eval(val) if isinstance(val, str) and val.startswith("[") else []
            # Normalize: lowercase, remove spaces within multi-word phrases (optional)
            skills.extend([skill.strip().lower().replace(" ", "_") for skill in items if isinstance(skill, str)])
        except:
            continue
    return ' '.join(skills)

df_rf["skills_text"] = df_rf.apply(combine_skills, axis=1)

# STEP 3: Vectorize all skills to compute importance
vectorizer_all = CountVectorizer(binary=True)
X_skills_all = vectorizer_all.fit_transform(df_rf["skills_text"])
y_rf = df_rf["SALARY"].fillna(df_rf["SALARY"].median())

# Train temp model to get skill importances
X_temp, _, y_temp, _ = train_test_split(X_skills_all, y_rf, test_size=0.3, random_state=42)
rf_temp = RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42)
rf_temp.fit(X_temp, y_temp)

# STEP 4: Select top 100 skills
all_skills = vectorizer_all.get_feature_names_out()
importances = rf_temp.feature_importances_
importance_df = pd.DataFrame({"Skill": all_skills, "Importance": importances})
top100_skills = importance_df.sort_values(by="Importance", ascending=False).head(100)["Skill"].tolist()

# STEP 5: Re-vectorize with only top 100
vectorizer_top100 = CountVectorizer(vocabulary=top100_skills, binary=True)
X_skills_100 = vectorizer_top100.fit_transform(df_rf["skills_text"])

# STEP 6: Train/test split (no STATE)
X_train_rf, X_test_rf, y_train_rf, y_test_rf = train_test_split(
    X_skills_100, y_rf, test_size=0.3, random_state=42
)
```

```python
# STEP 7: Set up and run RandomizedSearchCV
param_dist = {
    "n_estimators": randint(100, 300),
    "max_depth": randint(10, 30),
    "min_samples_split": randint(2, 10),
    "min_samples_leaf": randint(1, 10),
    "max_features": randint(10, X_skills_100.shape[1])
}

rf_model = RandomForestRegressor(random_state=42)
search_rf = RandomizedSearchCV(
    rf_model,
    param_distributions=param_dist,
    n_iter=15,
    scoring='neg_root_mean_squared_error',
    cv=3,
    verbose=1,
    random_state=42,
    n_jobs=-1
)

search_rf.fit(X_train_rf, y_train_rf)

# STEP 8: Evaluate model
best_model = search_rf.best_estimator_
y_pred_rf = best_model.predict(X_test_rf)

rmse = np.sqrt(mean_squared_error(y_test_rf, y_pred_rf))
r2 = r2_score(y_test_rf, y_pred_rf)

print("Best Hyperparameters:", search_rf.best_params_)
print(f"RMSE: {rmse:.2f}")
print(f"R Score: {r2:.3f}")
```
